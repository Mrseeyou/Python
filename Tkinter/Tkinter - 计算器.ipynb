# Tkinter - 计算器
- 模拟系统的计算功能
- 实现一个简单的具有加减法等操作的计算器
- 使用tkinter
- 操作步骤
    - 画GUI
    - 给每个控件配置相应的事件
    - 写逻辑代码

```python
# 第一步: 画出图形界面
from tkinter import *

root = Tk()
# 定义面板的大小
root.geometry('250x380')
root.title("计算器")

# 定义框架
# bg代表背景颜色(background), #dddddd是十六进制 表示颜色的一个串
frame_show = Frame(width=300, height=150, bg='#dddddd')
frame_show.pack()

# 定义顶部区域

sv = StringVar()
sv.set('0')

# anchor 表示的是锚点
# font 表示字体
show_label = Label(frame_show, textvariable=sv, bg='green',\
                   width=12, height=1, font=("黑体", 20, 'bold'), \
                   justify=LEFT,anchor='e')
show_label.pack(padx=10,pady=10)

# 第二步: 画出图形界面下半部分
# 按键区域
frame_bord = Frame(width=400, height=350, bg='#cccccc')

def delete():
    print("delete")
def clear():
    print("clear")
    
def fan():
    print("fan")
    
def ce():
    print('ce')    
   
# 定义最上一排的按钮
b_del = Button(frame_bord, text='←',width=5, height=1,\
               command=delete).grid(row=0, column=0)
button_ce = Button(frame_bord, text='CE', width=5, height=1,\
                  command=ce).grid(row=0, column=1)
button_clear = Button(frame_bord, text='C', width=5, height=1,\
                      command=clear).grid(row=0, column=2)
button_fan = Button(frame_bord, text='±', width=5, height=1,\
                   command=fan).grid(row=0, column=3)


# 定义数字0~9


b_7 = Button(frame_bord, text='7', width=5,height=2,\
             command=lambda:number('7')).grid(row=1,column=0)
b_8 = Button(frame_bord, text='8', width=5,height=2,\
             command=lambda:number('8')).grid(row=1,column=1)
b_9 = Button(frame_bord, text='9', width=5,height=2,\
             command=lambda:number('9')).grid(row=1,column=2)
b_4 = Button(frame_bord, text='4', width=5,height=2,\
             command=lambda:number('4')).grid(row=2,column=0)
b_5 = Button(frame_bord, text='5', width=5,height=2,\
             command=lambda:number('5')).grid(row=2,column=1)
b_6 = Button(frame_bord, text='6', width=5,height=2,\
             command=lambda:number('6')).grid(row=2,column=2)
b_1 = Button(frame_bord, text='1', width=5,height=2,\
             command=lambda:number('1')).grid(row=3,column=0)
b_2 = Button(frame_bord, text='2', width=5,height=2,\
             command=lambda:number('2')).grid(row=3,column=1)
b_3 = Button(frame_bord, text='3', width=5,height=2,\
             command=lambda:number('3')).grid(row=3,column=2)
b_0 = Button(frame_bord, text='0', width=5,height=2,\
             command=lambda:number('0')).grid(row=4,column=0)
b_10 = Button(frame_bord, text='.', width=5,height=2,\
             command=lambda:number('.')).grid(row=4,column=1)

# 定义 / * + - =

b_chu = Button(frame_bord, text='/', width=5,height=2,\
             command=lambda:operation('/')).grid(row=1,column=3)
b_cheng = Button(frame_bord, text='*', width=5,height=2,\
             command=lambda:operation('*')).grid(row=2,column=3)
b_jia = Button(frame_bord, text='+', width=5,height=2,\
             command=lambda:operation('+')).grid(row=3,column=3)
b_jian = Button(frame_bord, text='-', width=5,height=2,\
             command=lambda:operation('-')).grid(row=4,column=3)
b_deng = Button(frame_bord, text='=', width=5,height=2,\
             command=lambda:operation('=')).grid(row=4,column=2)

frame_bord.pack(padx=10,pady=10)

# 第三步: 逻辑, 即定义函数number和opertion
num1 = ''
num2 = ''
operator = ''
rst = ''

def number(num):
    global num1
    global num2
    global opertor
    
    if not operator:
        num1 = num1 + num
        # 如果是第一个操作数,则只显示第一个操作数
        sv.set(num1)
    else:
        num2 = num2 + num
        # 如果是第二个操作数,则显示完整的计算式子  
        sv.set(num2)
    
def operation(oper):
    global rst
    
    if oper in ['+', '-', '*', '/']:
        operator = oper
        sv.set(num1+oper+num2)
    else:
        if oper == '+':
            rst = int(num1) + int(num2)
        if oper == '-':
            rst = int(num1) - int(num2)
        if oper == '*':
            rst = int(num1) * int(num2)
        if oper == '/':
            rst = int(num1) / int(num2)
    
        sv.set(str(rst))
    
    
root.mainloop()
```
